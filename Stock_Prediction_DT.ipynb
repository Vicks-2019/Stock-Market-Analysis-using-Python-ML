```
# To predict the stock of the price will increase or decrease using machine learning's Decision Tree Algorithm
```

```
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas as pd
```

```
# load the data 
from google.colab import files
files.upload()
```

```
# storing data in dataframe
df = pd.read_csv('Book1.csv')
df
```

```
df = df.dropna()
df
```

```
# set date as index for data 
df = df.set_index(pd.DatetimeIndex(df['Date'].values))
df.index.name = 'Date'
df
```

```
# manipulate the data 
df['Price_up'] = np.where(df['Close'].shift(-1) > df['Close'],1,0)
# remove the date column from dataset
df = df.drop(columns=['Date'])
df
```

```
# split the dataset 
X = df.iloc[:,0:df.shape[1]-1].values
Y = df.iloc[:,df.shape[1]-1].values
```

```
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```
# create a model
tree = DecisionTreeClassifier().fit(X_train,Y_train)
```

```
print(tree.score(X_test,Y_test))
```

```
#show the model prediction
tree_prediction = tree.predict(X_test)
print(tree_prediction)
```
